```python
#Dictionary use
s={}#Empty dictionary
myDict={"Name":"Ritik Raushan","Age":21,"City":"Patna"}
print(myDict)


name=["Ritik", "Raushan","Sachin","Saurabh","Saloni"] #List

my_Dict=dict([('mame','Ritik'),('roll',18)])
mmy_Dict=dict({1:'Ritik',2:'Raushan'})

print(my_Dict)

#Access dictionary elements
print(my_Dict["mame"])
print(my_Dict.get('Age'))
my_Dict['Age']=22 #Updating the value of Age
print(my_Dict)

for i in my_Dict:
    print(i) #This will print the keys of the dictionary  

for i in my_Dict.values():
    print(i) #This will print the values of the dictionary

for i in my_Dict.items():
    print(i) #This will print the key-value pairs of the dictionary as tuples
```

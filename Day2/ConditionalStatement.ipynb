```python
c="start"

match c:
    case "start":
        print("Starting")
    case "stop":
        print("Stopping")
    case _:#Default case
        print("Invalid")
```

```python
c=51

match c:
    case x if x<0:
        print("Negative")
    case x if x%2==0:
        print("Even Number")
    case _:#Default case
        print( "Odd or Invalid or Other Number")
```

```python
#Pass Statement
#The pass statement is a null operation; it does nothing when executed.
#Pass is simply a placeholde but continue will skip the rest of the code for that iteration

num=[1,2,3,-4,-5,-6,-7,8,9]

print("Negative numbers")
for i in num:
    if i>0:
        pass 
    else:
        print(i)

print("Positive numbers")
for i in num:
    if i<0:
        pass 
    else:
        print(i)
```

```python
name=input("Enter your name: ")

for i in name:
    if(i=='t'):
        break
    print(i)

for i in name:
    if(i=='t'):
        continue
    print(i)


#Task: Print odd numbers from 1 to 10 using continue statement
for x in range(1, 10):
    if x%2==0:
        continue
    print(x)
```

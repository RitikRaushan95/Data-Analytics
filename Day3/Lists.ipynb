```python
p=input("Enter a password: ")
if len(p)<8:
    print("Weak password")
elif not any(ch.isdigit() for ch in p):
    print(" At least one digit is required in the password.")
elif not any(ch.isupper() for ch in p):
    print(" At least one uppercase letter is required in the password.")
elif not any(ch.islower() for ch in p):
    print(" At least one lowercase letter is required in the password.")
else:
    print("Strong password")
```

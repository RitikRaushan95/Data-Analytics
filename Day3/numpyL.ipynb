```python
import numpy as np
import time
import sys
SIZE=1000
l1=range(SIZE)
l2=range(SIZE)
a1=np.arange(SIZE)
a2=np.arange(SIZE)
start=time.time()
result=[(x,y) for x,y in zip(l1,l2)]
print("Time taken for list comprehension: ",(time.time()-start)*1000)
start=time.time()
result=a1+a2
print("Time taken for numpy array addition: ",(time.time()-start)*1000)
```

```python
import numpy as np
import time
start=time.time()
r=[]
l=list(range(1000000))

for i in l:
    r.append(i**2)
    e=time.time()
    
print("Time taken for list comprehension: ",(e-start))
```
